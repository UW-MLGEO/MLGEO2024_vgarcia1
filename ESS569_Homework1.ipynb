```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining the function to plot a graph
def plot(x, y, title, xlabel, ylabel):
    # Plotting the graph
    plt.plot(x, y)
    # Adding title to the graph
    plt.title(title)
    # Adding x-axis label
    plt.xlabel(xlabel)
    # Adding y-axis label
    plt.ylabel(ylabel)
    # Displaying the graph
    plt.show()

    plt.savefig('/Users/valeriagarcia/Desktop/ESS_569_ML_Geo/MLGEO2024_vgarcia1/ESS569_HW1_plot_cosine_function.png')
```

```python
# Define an array of x values from 0 to 10 with a spacing of 0.1
x = np.arange(0, 10, 0.1)

# Compute the y values as the cosine of x
y = np.cos(x)

# Define the x-axis and y-axis labels
plotting_labels = ['x', 'cos(x)']

# Run the plot function
plot(x, y, title = 'Plot of Cosine Function', xlabel = f'{plotting_labels[0]}', ylabel = f'{plotting_labels[1]}')
```
